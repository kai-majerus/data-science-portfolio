```python
import requests
import pandas as pd
import numpy as np
```

```python
api_key = "041733e1f80e420792d185f086958755"
base_url = 'https://api.football-data.org/v4'
headers = {'X-Auth-Token': api_key}
```

```python
# Function to get historical matches for a team
def get_team_matches(team_id, season='2024'):
    url = f'{base_url}/teams/{team_id}/matches?season={season}'
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()['matches']
    else:
        print(f"Error: {response.status_code}")
        return None


# Example: Collect matches for Liverpool (team ID = 64) for the 2024 season
team_id = 64  # Liverpool's team ID
matches = get_team_matches(team_id, season='2024')

# Parse and store the relevant match data into a DataFrame
match_data = []
for match in matches:
    match_data.append({
        'date': match['utcDate'],
        'home_team': match['homeTeam']['name'],
        'away_team': match['awayTeam']['name'],
        'home_score': match['score']['fullTime']['home'],
        'away_score': match['score']['fullTime']['away'],
        'status': match['status']
    })

df_matches = pd.DataFrame(match_data)
print(df_matches)
```
